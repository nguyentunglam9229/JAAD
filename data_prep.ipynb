```python
import numpy as np
def resize_annotations_bbox(w, h, annotations_data):
    data = annotations_data.copy()
    for video_name in data.keys():
        v_width = data[video_name]['width']
        v_height = data[video_name]['height']
        w_ratio = w/v_width
        h_ratio = h/v_height
        ped_annotations = data[video_name]['ped_annotations']
        for k in ped_annotations.keys():
            bbox = np.array(ped_annotations[k]['bbox'])
            bbox[:, 0] = bbox[:, 0]*w_ratio
            bbox[:, 1] = bbox[:, 1]*h_ratio
            bbox[:, 2] = bbox[:, 2]*w_ratio
            bbox[:, 3] = bbox[:, 3]*h_ratio
            ped_annotations[k]['bbox'] = bbox.tolist()
    return data

data_resized = resize_annotations_bbox(480, 270, data)
```

```python
def save_annotations(annotations, path):
    for i in range(len(annotations)):
        with open(f'{path}/frame{i:05d}.txt', 'w') as f:
            for j in annotations[i]:
                f.write(f'{j}\n')
```
